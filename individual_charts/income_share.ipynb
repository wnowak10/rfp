```python
import pandas as pd
df = pd.read_csv('../Desktop/inc.tsv', sep = '\t')
df.sort_values(by = 'Year', ascending = True, inplace = True)
df['Year'] = df['Year'].astype(str)
df.dropna(inplace = True)
df.set_index('Year', inplace = True, drop = False)
```

```python
df['bot_90_perc'] = (df['Bottom 90'] / 100) * df['GDP']
```

```python
bot90max = df['Bottom 90'].max()
bot90min = df['Bottom 90'].min()
```

```python
import numpy as np
df['color'] = np.where(df['Bottom 90'] == bot90max, 'g', 
                      np.where(df['Bottom 90'] == bot90min, 'r', 'b'))
# df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")

# Initialize the matplotlib figure
# f, ax = plt.subplots(figsize=(6, 15))

fig = plt.figure()
fig.set_size_inches(12, 15, forward = True)
# plt.figure(figsize=(50,50))

fig.suptitle('Real GDP, 90/10 income over time', fontsize=14, fontweight='bold')

ax = fig.add_subplot(111)

# Load the example car crash dataset
# crashes = sns.load_dataset("car_crashes").sort_values("total", ascending=False)



# for x in df['Bottom 90']]
# c = ['b' if x > df['Bottom 90'].min() else 'r' for x in df['Bottom 90']]


sns.set_color_codes("pastel")
sns.barplot(x="GDP", y="Year", 
            order = df['Year'], 
            data=df, label="Income of top 10th Percentile", 
            palette=df['color'], orient = 'h')


# Plot the crashes where alcohol was involved
sns.set_color_codes("muted")
sns.barplot(x="bot_90_perc", y="Year", data=df,order = df['Year'],
            label="Income of bottom 90th Percentile", palette=df['color'], orient = 'h')

# Add a legend and a informative axis label
ax.legend(ncol=2, loc="lower right", frameon=True)
# ax.set(xlim=(0, 24), ylabel="",
#        xlabel="Automobile collisions per billion miles")
ax.set(xlabel = 'Real GDP')

ax.annotate('In 2012, the top 10% of the income distribution took home 48% of national income.', xy=(.2, .95),  xycoords='axes fraction')
ax.annotate('In 1970, the top 10% of the income distribution took home only 34% of national income.', xy=(.2, .157),  xycoords='axes fraction')


# ,
#             xytext=(0.8, 0.95), textcoords='axes fraction',
#             arrowprops=dict(facecolor='black', shrink=0.05),
#             horizontalalignment='right', verticalalignment='top',
#             )
sns.despine(left=True, bottom=True)
plt.show()
fig.savefig('test3png.png', dpi=200)
```
